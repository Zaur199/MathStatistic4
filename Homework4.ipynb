Задача 1. Случайная непрерывная величина A имеет равномерное распределение на
промежутке (200, 800].
Найдите ее среднее значение и дисперсию.

РЕШЕНИЕ

- Среднее значение равно: (a+b)/2 =(200+800)/2 = 500.0
- Дисперсия равняется: (b-a)^2/12 =(800-200)^2/12= 30000.0

Задача 2. О случайной непрерывной равномерно распределенной величине B известно, что ее
дисперсия равна 0.2.
Можно ли найти правую границу величины B и ее среднее значение зная, что левая
граница равна 0.5?
Если да, найдите ее.

РЕШЕНИЕ

Используем формулу нахождения дисперсии, подставляя известные параметры. В результате имеем:

- D(X) = (b-a)**2/12
- 0.2 = (b-0.5)**2/12
- (12 * 0.2) ** 0.5 = b - 0.5
- b = ((12 * 0.2) ** 0.5) + 0.5
- В результате решая полученное уравнение получим:
- b = 2.05(правая граница)
- среднее значение: (0.5 + b)/2 = 1.27

Задача 5. На сколько сигм (средних квадратичных отклонений) отклоняется рост человека,
равный 190 см, от
математического ожидания роста в популяции, в которой M(X) = 178 см и D(X) = 25 кв.см?

РЕШЕНИЕ

- Дисперсия равна: 5 см
- Количество сигм(std(X)) находим по формуле: X-M(X)/std(X) = (190 - 178)/5 = 2.4

Задача 4. Рост взрослого населения города X имеет нормальное распределение, причем, 
средний рост равен 174 см, а среднее 
квадратическое отклонение равно 8 см. посчитайте, какова вероятность того, 
что случайным образом выбранный взрослый человек имеет рост:

1. больше 182 см?
2. больше 190 см?
3. от 166 см до 190 см?
4. от 166 см до 182 см?
5. от 158 см до 190 см?
6. не выше 150 см или не ниже 190 см?
7. не выше 150 см или не ниже 198 см?
8. ниже 166 см?

Задачу можно решить двумя способами: без использования сторонних библиотек (numpy, scipy, pandas и пр.), 
а затем проверить себя с помощью встроенных функций
- РЕШЕНИЕ

```python
from scipy import stats
```

```python
P1 = stats.norm.cdf(182, loc=174, scale=8)
print(f'1. Вероятность того что взрослый человек имеет рост больше 182 см: 1-P1 = {round((1-P1),3)}')
P2 = stats.norm.cdf(190, loc=174, scale=8)
print(f'2. Вероятность того что взрослый человек имеет рост больше 190 см: 1-P2 = {round((1-P2),3)}')
P3 = stats.norm.cdf(166, loc=174, scale=8)
print(f'3. Вероятность того что взрослый человек имеет рост от 166 см до 190 см: P2-P3 = {round((P2-P3),3)}')
print(f'4. Вероятность того что взрослый человек имеет рост от 166 см до 182 см: P1-P3 = {round((P1-P3),3)}')
P4 = stats.norm.cdf(158, loc=174, scale=8)
print(f'5. Вероятность того что взрослый человек имеет рост от 158 см до 190 см: P2-P4 = {round((P2-P4),3)}')
P5 = stats.norm.cdf(150, loc=174, scale=8)
print(f'6. Вероятность того что взрослый человек имеет рост не выше 150 см или не ниже 190 см: 1-(P2-P5) = {round((1-(P2-P5)),3)}')
P6 = stats.norm.cdf(198, loc=174, scale=8)
print(f'7. Вероятность того что взрослый человек имеет рост не выше 150 см или не ниже 198 см: 1-(P6-P4) = {round((1-(P6-P5)),4)}')
print(f'8. Вероятность того что взрослый человек имеет рост ниже 166 см: 1-P3 = {round((1-P3),3)}')
```
